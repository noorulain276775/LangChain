```python
from statistics import correlation

import kagglehub
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f'{path}/books.csv')
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), ax=ax)
plt.xlabel('Columns')
plt.ylabel('Missing Values')
plt.show()
```

```python
import numpy as np
books['missing_description'] = np.where(books['description'].isna(), 1, 0)
books['age_of_the_book'] = 2025 - books['published_year']
```

```python
columns_of_interest = ["num_pages", "age_of_the_book", "missing_description", "average_rating"]
```

```python
correlation_matrix = books[columns_of_interest].corr(method= 'spearman')
sns.set_theme(style="white")
plt.figure(figsize=(10,10))
myHeatmap = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt='.2f', cbar_kws={"label": "Spearman Correlation"})
myHeatmap.set_title('Correlation Heatmap')
plt.show()


# So can not see any strong relation of missing book_description with other values which can bias our results
```

```python
refined_books = books[~(books['description'].isna()) &
      ~(books['num_pages'].isna()) &
      ~(books['published_year'].isna()) &
      ~(books['average_rating'].isna())
]

# Result is under 95% so we can delete 5% of missing data
```

```python
# Now in our data we can see we have 531 different categories and some of them are so specific that it is useless so lets just observe these categories

refined_books['categories'].value_counts().reset_index().sort_values('count', ascending=False)

# From chart below we can see we have problem with the categories of this dataset
```

```python
# the top result is uneven distribution so let's observe the data more closely
refined_books = (
    refined_books
      .copy()
      .assign(words_in_description=lambda df: df['description'].str.split().str.len())
)
```

```python
# make a real copy of the filtered rows
more_refined_books = refined_books.loc[refined_books['words_in_description'] >= 25].copy()

# build title_and_subtitle safely
more_refined_books['title_and_subtitle'] = np.where(
    more_refined_books['subtitle'].isna(),
    more_refined_books['title'].astype(str),
    more_refined_books['title'].astype(str) + ': ' + more_refined_books['subtitle'].astype(str)
)
```

```python
more_refined_books
```

```python
more_refined_books["tagged_description"] = more_refined_books[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
(
    more_refined_books
    .drop(["subtitle", "missing_description", "age_of_the_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index = False)
)
```
